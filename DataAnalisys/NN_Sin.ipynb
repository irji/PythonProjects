```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

values = {}

#base_v=0
num_of_points=200
#step=

for x in range(num_of_points + 1):
    #values[base_v + 0.1*x]=np.sin(base_v + 0.1*x)
    b1=20*np.random.rand()
    values[b1]=np.sin(b1)

#print(values)

plt.figure(figsize=(15,5))

plt.plot(values.keys(), values.values(), '.', color='blue')

plt.xlabel('x values')
plt.ylabel('y values')
plt.title('sin plot')
#plt.grid(True)
plt.show()
```
