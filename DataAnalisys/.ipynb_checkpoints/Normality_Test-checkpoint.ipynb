Для справки использовал статью https://plot.ly/python/normality-test/ и документацию к scipy (https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kstest.html#scipy.stats.kstest).

Так же есть информация в wiki (https://en.wikipedia.org/wiki/Normal_probability_plot и https://en.wikipedia.org/wiki/Shapiro%E2%80%93Wilk_test и https://machinelearningmastery.com/a-gentle-introduction-to-normality-tests-in-python/)

Для расчета в excel есть хорошая статья https://www.spcforexcel.com/knowledge/basic-statistics/normal-probability-plots

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

myFile = pd.read_csv('text.txt', sep=',')[13503:]

xxx=[]
yyy=[]
zzz1=[]
zzz2=[]
zzz3=[]

#fig = plt.subplots(nrows=2)
#fig, axes = plt.subplots(nrows=2)

fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

for x1 in range(1,11):
    for y1 in range(1,11):

        basis = (x1 * myFile['SPFB.GAZR'] - y1 * myFile['SPFB.VTBR']).values

        ks_results = stats.kstest(basis, cdf='norm')
        anderson_results = stats.anderson(basis)
        dagostino_results = stats.normaltest(basis)

        xxx.append(x1)
        yyy.append(y1)
        zzz1.append(ks_results[1])
        zzz2.append(anderson_results[1][2])
        zzz3.append(dagostino_results[1])
        
        df_norm = (basis - basis.mean()) / (basis.max() - basis.min())
        #df_norm = (basis - basis.min()) / (basis.max() - basis.min())
        
        #ax1 = plt.subplot(2,1,2)
        ax1.plot(df_norm)

        #axes[0].plot(basis)

        #if ks_results[1] > 0.1:
        #    print(str(x1) + "," + str(y1))
        #if anderson_results[1][2] > 0.1:
        #    print(str(x1) + "," + str(y1))
        #if dagostino_results[1] > 0.1:
        #    print(str(x1) + "," + str(y1))

df=pd.DataFrame({'X1': xxx,'Y1': yyy, 'Kolmogorov_Smirnov': zzz1, 'Anderson_Darling': zzz2, 'DAgostino_and_Pearson': zzz3})

#print(df)

#plt.figure(figsize=(15,5))
#plt.subplot(2,1,2)
#plt.plot(df['Kolmogorov_Smirnov'], color='blue')
#plt.plot(df['Anderson_Darling'], color='green')
#plt.plot(df['DAgostino_and_Pearson'], color='red')

#ax2 = plt.subplot(2,1,1)
ax2.plot(df['Kolmogorov_Smirnov'], color='blue')
ax2.plot(df['Anderson_Darling'], color='green')
ax2.plot(df['DAgostino_and_Pearson'], color='red')

#axes[1].plot(df['Kolmogorov_Smirnov'], color='blue')
#axes[1].plot(df['Anderson_Darling'], color='green')
#axes[1].plot(df['DAgostino_and_Pearson'], color='red')

plt.show()
```
